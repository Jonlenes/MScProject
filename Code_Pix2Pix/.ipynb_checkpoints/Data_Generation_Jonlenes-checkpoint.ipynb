```python
import time
import os
from random import random
import multiprocessing as mp
from PIL import Image
import glob
import matplotlib.pyplot as plt
from matplotlib import patches, patheffects
import numpy as np
import cv2
import random

global_pre_path = "datasets/jonlenes"
```

```python
def show_img(img, show_axis=False):
    if not show_axis: plt.axis("off")
    plt.imshow(img)
    plt.show()

def open_image(path, color_code=cv2.COLOR_BGR2RGB):
    return cv2.cvtColor(cv2.imread(path), color_code) 
    
```

```python
def f(x):
    y = 0 # random.randint(20, 256)
    result = []
    for _ in x:
        result.append(y)
        y += np.random.normal(scale=1)
    return np.array(result)

def runningMean(x, N):
    return np.convolve(x, np.ones((N,))/N)[(N-1):]
```

```python
def make_sample(index, pre_path=global_pre_path):
    list_curves = []
    n_curves = random.randint(5, 10)

    for i in range(n_curves):
        n_points = random.randint(20, 250)
        n_smooth = random.randint(20, 40)
        x_start = random.randint(1, 256 - n_smooth)

        x = np.linspace(1, n_points) + x_start
        y = runningMean(f(x), n_smooth)

        list_curves.append((x, y))
    
    names = ["x", "y"]
    form = ["", "--"]
    fig, ax = plt.subplots()
    for i in range(2):
        ax.axis('off')
    
        for x, y in list_curves:
            ax.plot(x, y, form[i])
        plt.savefig("./" + pre_path + "/" + names[i] + '_' + str(index) + '.jpg', bbox_inches='tight', pad_inches=-0.2)
        plt.cla()
    plt.close(fig)
```

```python
make_sample(0)
img_x = cv2.resize(open_image(global_pre_path + '/x_0.jpg'), (256, 256))
img_y = cv2.resize(open_image(global_pre_path + '/y_0.jpg'), (256, 256))
print(img_x.shape, img_y.shape)
show_img(img_x)
show_img(img_y)
```

```python
from scipy.signal import convolve2d
from scipy.signal import ricker

def apply_ricker(x):
    vec_ricker = ricker(10, 4).reshape(1, 10) 
    x = convolve2d(x, vec_ricker)
    return x
```

```python
from skimage import img_as_ubyte

def adjust_sample(index, path=global_pre_path):
    x = cv2.resize(open_image(path + '/x_' + str(index) + '.jpg'), (256, 256))
    y = cv2.resize(open_image(path + '/y_' + str(index) + '.jpg'), (256, 256))
    
    os.remove(path + '/x_' + str(index) + '.jpg')
    os.remove(path + '/y_' + str(index) + '.jpg')
    
    sample = np.append(y, x, axis=1)
    vmin, vmax = sample.min(), sample.max() 
    sample = (sample - vmin) / (vmax - vmin)
    
    cv2.imwrite(path + '/sample_' + str(index) + '.jpg', sample * 255)
```

```python
make_sample(0)
adjust_sample(0)
sample = open_image(global_pre_path + '/sample_0.jpg')
print(sample.shape)
show_img(sample)
```

```python
def create_dataset(n_samples, pre_path=global_pre_path):
    folders = ["train", "val", "test"]
    list_n_samples = [n_samples, int(n_samples * 0.2), int(n_samples * 0.2)]
    
    for j in range(3):
        print("Generating: ", folders[j])
        path = pre_path + "/" + folders[j]
        if not os.path.exists(path):
            os.mkdir(path)
        for i in range(list_n_samples[j]):
            make_sample(i, path)
            adjust_sample(i, path)
            if (i  + 1) % 100 == 0:
                print("Percent: {0}%".format(int((i / n_samples) * 100)))
```

```python
create_dataset(1000)
```

```python
def get_gauss_kernel(size=3,sigma=1):
    center=(int)(size/2)
    kernel=np.zeros((size,size))
    for i in range(size):
        for j in range(size):
            diff=np.sqrt((i-center)**2+(j-center)**2)
            kernel[i,j]=np.exp(-(diff**2)/2*sigma**2)
    return kernel/np.sum(kernel)

# psf = get_gauss_kernel(5, 1)
# img_res = convolve2d(img, psf)
# show_img(img_res)
```
